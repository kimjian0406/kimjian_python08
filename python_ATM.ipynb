```python
# 초기 데이터: 계좌 정보
accounts = {
    "123-456-789": {"비밀번호": "1234", "잔액": 100000, "거래내역": []},
    "987-654-321": {"비밀번호": "5678", "잔액": 50000, "거래내역": []},
}

def 로그인():
    """사용자 로그인 기능"""
    print("\nATM 시스템에 오신 것을 환영합니다!")
    while True:
        account_number = input("계좌 번호를 입력하세요: ")
        if account_number in accounts:
            password = input("비밀번호를 입력하세요: ")
            if accounts[account_number]["비밀번호"] == password:
                print("\n로그인 성공!")
                return account_number
            else:
                print("비밀번호가 올바르지 않습니다.")
        else:
            print("계좌 번호가 존재하지 않습니다.")

def 잔액조회(account_number):
    """잔액 조회 기능"""
    print(f"\n현재 잔액: {accounts[account_number]['잔액']}원")

def 입금(account_number):
    """입금 기능"""
    try:
        amount = int(input("입금할 금액을 입력하세요: "))
        if amount > 0:
            accounts[account_number]["잔액"] += amount
            accounts[account_number]["거래내역"].append(f"입금: {amount}원")
            print(f"{amount}원이 입금되었습니다.")
        else:
            print("0원 이상을 입력하세요.")
    except ValueError:
        print("올바른 금액을 입력하세요.")

def 출금(account_number):
    """출금 기능"""
    try:
        amount = int(input("출금할 금액을 입력하세요: "))
        if 0 < amount <= accounts[account_number]["잔액"]:
            accounts[account_number]["잔액"] -= amount
            accounts[account_number]["거래내역"].append(f"출금: {amount}원")
            print(f"{amount}원이 출금되었습니다.")
        elif amount > accounts[account_number]["잔액"]:
            print("잔액이 부족합니다.")
        else:
            print("0원 이상을 입력하세요.")
    except ValueError:
        print("올바른 금액을 입력하세요.")

def 거래내역조회(account_number):
    """거래 내역 조회 기능"""
    print("\n[거래 내역]")
    if accounts[account_number]["거래내역"]:
        for record in accounts[account_number]["거래내역"]:
            print(record)
    else:
        print("거래 내역이 없습니다.")

def ATM():
    """ATM 메인 함수"""
    account_number = 로그인()
    while True:
        print("\n[ATM 메뉴]")
        print("1. 잔액 조회")
        print("2. 입금")
        print("3. 출금")
        print("4. 거래 내역 조회")
        print("5. 종료")
        choice = input("원하는 작업을 선택하세요: ")

        if choice == "1":
            잔액조회(account_number)
        elif choice == "2":
            입금(account_number)
        elif choice == "3":
            출금(account_number)
        elif choice == "4":
            거래내역조회(account_number)
        elif choice == "5":
            print("시스템을 종료합니다. 이용해주셔서 감사합니다!")
            break
        else:
            print("올바른 선택이 아닙니다. 다시 입력해주세요.")

# 프로그램 실행
ATM()
```
